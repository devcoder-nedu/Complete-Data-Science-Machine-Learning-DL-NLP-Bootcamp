### Assignment 8: break Statement

Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

```python
sum = 0
while True:
    number = int(input("Please input numbers except zero unless you want to stop the program"))
    if number == 0:
        break
    sum += number

print(sum)
```

### Assignment 12: Factorial Calculation

Write a program that calculates the factorial of a number input by the user using a while loop.

```python
user_input = int(input("Please input a factorial"))

number = user_input
n = 1
while number > 1:
    n *= number
    number -= 1

print(n)

    
```

### Assignment 13: Sum of Digits

Write a program that calculates the sum of the digits of a number input by the user using a while loop.

```python
user_input = (input("Please input the number whose digits you wish to sum"))
length = len(user_input)
sum = 0
while length > 0:
    sum += int(user_input[length -1 ])
    length -= 1

print(sum)

```

### Assignment 14: Prime Number Check

Write a program that checks if a number input by the user is a prime number using a for loop.


```python
user_input = int(input("Please, input a number to check if it is prime or not"))

if user_input == 2:
    print("2 is a prime")
else:
    n = 0
    for x in range (2, user_input):
        if user_input % x == 0:
            print(f"{user_input} is not a prime")
            n += 1
            break

    if n == 0:
        print(f"{user_input} is a prime")
```
